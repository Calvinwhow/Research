# Get Scatterplot

```python
from scipy.stats import spearmanr, pearsonr
from natsort import index_natsorted
import numpy as np
import seaborn as sns
import os
def generate_scatterplot(dataframe, columns_per_row=4, dependent_variable_name='axis_name', independent_variable_name='axis_name', independent_variable_units='axis_units',
                         x_label='xlabel', y_label='ylabel'):
    '''
    This will gerenate scatterplots with SpearmanR annotated
    
    Input: 
    dataframe - just the parts of the dataframe you wish to plot. 
    
    Returns:
    figure
    '''
    #Melt the data frame
    melted_df = pd.melt(dataframe, 
                        id_vars=[dependent_variable_name], value_vars=dataframe.columns.values.tolist()[:],
                        var_name=independent_variable_name, value_name=independent_variable_units
                        );
    # sort the DataFrame by the Connectivity Region column
    melted_df = melted_df.sort_values(by=independent_variable_name, key=lambda x: np.argsort(index_natsorted(melted_df[independent_variable_name])))
    display(melted_df)
    
    #Begin plotting 
    sns.set_style('white')
    sns.set_palette('Greys', 1, desat=1)
    facet_plot = sns.lmplot(y=dependent_variable_name, x=independent_variable_units, data=melted_df,
            col=independent_variable_name,
            truncate=False, col_wrap=columns_per_row,
            facet_kws=dict(sharex=False, sharey=True)) #sharex and sharey disallows axis from ranging across subplots. Good for visualizaiton. 

    #Calculate pearsonr for each cluster of Connection Regions
    grouped_df = melted_df.groupby(independent_variable_name)
    result = grouped_df.apply(lambda x: spearmanr(x[independent_variable_units], x[dependent_variable_name]))


    #Organize the outputs of the pearson to match the facet plot
    rearranged_results = []
    for region in melted_df[independent_variable_name].unique():
        rearranged_results.append(result[region])
    result = rearranged_results
        
    for i, ax in enumerate(facet_plot.axes):
        r, p = result[i]
        region = melted_df[independent_variable_name].unique()[i]
        title = f"{region}"
        ax.set_title(title)
        ax.annotate(f"r = {r:.2f}, p = {p:.5f}", xy=(.5, 1.0), xycoords='axes fraction',
                xytext=(0, 0), textcoords='offset points',
                ha='center', va='top',
                bbox=dict(boxstyle='round,pad=0.0', alpha=0.0),
                arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
        
    facet_plot.set_xlabels(x_label)
    facet_plot.set_ylabels(y_label)
    return facet_plot
```

```python
independent_variable = 'Non-Machine Learning'
dependent_variable = 'ACE3_Other'
#----------------------------------------------------------------
columns_of_interest = [independent_variable, dependent_variable]
scatterplot = generate_scatterplot(data_df.loc[:, columns_of_interest], dependent_variable_name=dependent_variable, independent_variable_name=independent_variable, independent_variable_units='Score',
                                   y_label='Paper Score', x_label='Autonomous Score')

#Save the Scatterplot Above
save_dirsvg = os.path.join(out_dir, f'linear_model_facet_plot_{independent_variable}.svg')
save_dirpng = os.path.join(out_dir, f'linear_model_facet_plot_{independent_variable}.png')
scatterplot.savefig(save_dirsvg)
scatterplot.savefig(save_dirpng)
print(f'Fig saved to ', save_dirpng)
```
