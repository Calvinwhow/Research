```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
#USER INPUT REQUIRED
file_dir = r'C:\Users\calvin.howard\Dropbox (Partners HealthCare)\memory\analyses\spcorrel\vtafxconn-yeofxconn/'
file_ending = '*.csv'
```

```python
print('path: ', os.path.join(file_dir, file_ending))
globbed = glob.glob(os.path.join(file_dir, file_ending))

collected_df = pd.DataFrame()
for i in range(0, len(globbed)):
    name = os.path.basename(globbed[i]).split('.csv')[0]
    if collected_df.empty:
        arb_df = pd.read_csv(globbed[i])#, names=name)
    else:
        arb_df = pd.read_csv(globbed[i],usecols=[1])# names=name, )
    collected_df = pd.concat([collected_df, arb_df], axis=1)
collected_df = collected_df.set_index('roi')
collected_df.tail(1)
```
