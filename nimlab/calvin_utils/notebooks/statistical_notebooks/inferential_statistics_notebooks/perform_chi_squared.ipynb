## Chi Squared Test Time

```python
## Choose variables of interest.
x_name = '00_memor'
z_name = '% Change from baseline (ADAS-Cog11)' #--------------------------------------------------------This variable is the outcome variable

x_vals = total_df[x_name]
z_vals = total_df[z_name]#------------------------------------------------------------------------------This variable is the outcome variable


#subgrouping by age
metric = 'Age at DOS'
metric_val = 65

index_one = (total_df[metric] > metric_val) #example, all individuals over 65
index_two = (total_df[metric] <= metric_val) #example, all individuals under/equal to 65

##Example of how to use these indices to manipulate data:
#### clin_df.loc[index_one, '<names of relevant columns>'] = 0 ## can use iloc too. 
```

```python
## BIN THE DATA BY PROBABILITY
import scipy.stats as st

x_zscore = st.zscore(x_vals)
y_zscore = st.zscore(z_vals)

z1 = st.norm.ppf(.25); z2 = st.norm.ppf(.5); z3 = st.norm.ppf(.75); z4 = st.norm.ppf(0.99999)
print(z1, z2, z3, z4)

#Prepare the datframe from contingency table creation
xbins = (x_zscore<=z2); ybins = (y_zscore<=z2)
xbins[xbins == True] = '<=z2'; ybins[ybins == True] = '<=z2'
xbins[xbins == False] = '>z2'; ybins[ybins == False] = '>z2'
total_df['xbins'] = xbins; total_df['ybins'] = ybins

#CALCULATE MARGINS OF THE CONTINGENCY TABLE
xbin1 = np.count_nonzero(x_zscore<=z2); xbin2 = np.count_nonzero(x_zscore >z2); #xbin3 = np.count_nonzero((x_zscore>z2) & (x_zscore<=z3)); xbin4 = np.count_nonzero(x_zscore>z3)
ybin1 = np.count_nonzero(y_zscore<=z2); ybin2 = np.count_nonzero(y_zscore >z2); # ybin3 = np.count_nonzero((y_zscore>z2) & (y_zscore<=z3)); ybin4 = np.count_nonzero(y_zscore>z3)

#CALCULATE OBSERVED VALUES OF THE CONTINGENCY TABLE
or1c1 = np.count_nonzero(total_df.index[(total_df['xbins'] == '<=z2') & (total_df['ybins'] == '<=z2')])+1
or2c1 = np.count_nonzero(total_df.index[(total_df['xbins'] == '>z2') & (total_df['ybins'] == '<=z2')])
or1c2 = np.count_nonzero(total_df.index[(total_df['xbins'] == '<=z2') & (total_df['ybins'] == '>z2')])
or2c2 = np.count_nonzero(total_df.index[(total_df['xbins'] == '>z2') & (total_df['ybins'] == '>z2')])
print(or1c1, or2c1, or1c2, or2c2) #, np.sum(r1c1, r2c1, r1c2, r2c2))

#CALCULATE THE EXPECTED VALUES OF THE CONTINGENCY TRABLE
er1c1 = xbin1*ybin1/len(total_df.index); 
er2c1 = xbin2*ybin1/len(total_df.index); 
er1c2 = xbin1*ybin2/len(total_df.index); 
er2c2 = xbin2*ybin2/len(total_df.index)
print(er1c1, er2c1, er1c2, er2c2)

cont_table_df = pd.DataFrame({'<=50% outcomes by prob': [or1c1, or2c1],
                            '>50% outcomes by prob': [or1c2, or2c2]}, index=['<=50% connect by prob', '>50% connect by prob'])

exp_table_df = pd.DataFrame({'<=50% outcomes by prob': [er1c1, er2c1],
                            '>50% outcomes by prob': [er1c2, er2c2]}, index=['<=50% connect by prob', '>50% connect by prob'])

##----------------------------------------------------------------
#THE ERROR IN THIS CODE IS THAT i NEED TO CALCULATE THE BINS FOR THE INTERSECTIONS OF EACH DATA GROUP
#SO, I NEED THE 'LOW PROB CONN AND HIGH PROB OUTCOME' BIN COUNT, ETC, FOR EACH. 
#NOW, WITH MY LOW LUMBERS, THIS CREATES 16 BINS, AND 50/16 = ~3.7, WHICH IS <5, THEREFOR CHI IS NOT APPLICABLE AND FISHER IS REQUIRED. 
#CAN CHANGE TO BING 4X2, AND 50/8=~6.25, WHICH IS STILL AT RISK OF <5 IN LOW/LOW PROB BINS, MEANING FISHER IS REQUIRED STILL.
#CAN CHANGE BIN TO 2X2, AND 50/4 = 12.5, WHICH IS MUCH SAFER. 
##----------------------------------------------------------------


#Bin by Quantile
# quantile = 0.25
# outcome = np.quantile(total_df[z_name], quantile)
# conn = np.quantile(total_df[x_name], quantile)
display(cont_table_df)
```
